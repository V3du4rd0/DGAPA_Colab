# ***Introducción a Colab***

Los archivos de Colab, cuya extensión es ipynb, los denominamos **notebooks**. En un notebook podemos disponer de dos tipos básicos de celdas:


*   _Celdas de texto_   (Text cells)
*   _Celdas de código_   (Code cells)



Las primeras sirven para dar soporte documental al contenido que estamos elaborando, asímismo nos podemos apoyar de algunas herramientas para dar formato, por ejemplo:

**Texto en negritas**

*Texto en itálicas*

~~Texto tachado~~

`Fuente monospace`

También disponemos de viñetas:

1.   List item
2.   List item

ó

*   List item
*   List item

Por otra parte, las celdas de Código sirven principalmente para ejecutar instrucciones en lenguaje Python. Sin embargo, como mostraremos en las siguientes unidades, podemos extender sus funciones más allá de Python.

Tenemos diferentes opciones para agregar celdas, por ejemplo:

* Desde el menú superior:

 Insert > Text cell / Code cell

* Si al final de una celda posicionamos el cursor aproximadamente a la mitad de pantalla, veremos que aparecen dos botones, uno con la leyenda **+ Code** y otro con **+ Text**. Al dar click sobre alguno de estos botones crearemos una nueva celda inmediatamente.

* Atajo: Podemos crear una celda de código con el siguiente atajo desde el teclado: **CTRL**+ **M B**

Una vez que escribimos el contenido de una celda de texto, debemos presionar **Shift + Enter** para ver los cambios integrados al documento.

También podemos volver a editar una celda dando doble click sobre ella.

Por otra parte, en celdas de código, además de **Shift + Enter** para ejecutar las instrucciones de Python, podemos de forma equivalente dar click en un ícono de Play que aparece cuando posicionamos el cursor en el extremo izquierdo de la celda. Veamos:


```python
1+1
```

# Algunas opciones de formato

Títulos y subtítulos para identificar secciones


El símbolo # al comienzo de la línea en una celda de texto nos permite señalar que estamos por escibir el nombre de un encabezado. También podemos incluir más de un # para encabezados de secciones o subsecciones, por ejemplo:

# Encabezado 1
## Encabezado 2
### Encabezado 3
#### Encabezado 4

# <font color='red'> Cambios de color</font>

El nombre del lenguaje que nos permite dar formato al texto dentro de Colab es Markdown, y aunque en general es sencillo de utilizar porque sólo se basa en patrones de caracteres, es muy limitado.

Sin embargo, podemos agregar funcionalidades con sintaxis de html, por ejemplo para cambiar el color:

<font color='blue'>Texto en color azul!</font>


# Cambio de tamaño de fuente

De una forma similar al cambio de color, podemos cambiar el tamaño de la fuente:

<font size=2>Texto de diferente tamaño!!</font>


# Saltos de línea

Si necesitamos añadir uno o más saltos de línea, podemos hacerlo de la misma manera que si utilizaramos HTML, es decir con **< br>**

Veamos un ejemplo:

Linea 1<br>Linea 2
